```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib import animation
import numpy as np
```

```python
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
#list of leagues
leagues = os.listdir("Data")
#list of seasons
seasons = os.listdir(os.path.join("Data",leagues[0]))
#list of tables
tables = os.listdir(os.path.join(*["Data", leagues[0], seasons[0]]))
```

```python
tables
```

```python
#names of data files
Defense_datafile = 'stage-team-stats-defensive.csv'
Action_side_datafile = 'stage-touch-channels.csv'
Action_zone_datafile = 'stage-touch-zones.csv'
Summary_datafile = 'stage-team-stats-summary.csv'
```

```python
aio = pd.DataFrame(columns = ['Team','OVR_Rating','Shots pg','Tackles pg','Interceptions pg','Fouls pg','Offsides pg','Defense_Rating','Left Side','Middle of the pitch','Right Side','Own Third','Middle Third','Opposition Third','League','Season'])
for league in leagues:
    for season in seasons:
        ranking = pd.read_csv(os.path.join(*['Data', league,season,Summary_datafile]))[['Team', 'Rating']]
        ranking['Team'] = ranking['Team'].apply(lambda x: ' '.join(x.split()[1:]))
        ranking = ranking.rename(columns={'Rating': 'OVR_Rating'})
        
        defense = pd.read_csv(os.path.join(*['Data', league,season,Defense_datafile]))
        defense['Team'] = defense['Team'].apply(lambda x: ' '.join(x.split()[1:]))
        defense = defense.rename(columns={'Rating': 'Defense_Rating'})
        
        action_side = pd.read_csv(os.path.join(*['Data', league,season,Action_side_datafile])).drop(columns = ['R'])
        
        
        action_zone = pd.read_csv(os.path.join(*['Data', league,season,Action_zone_datafile])).drop(columns = ['R'])
        
        
        full = pd.DataFrame(columns = ['Team'])
        full = pd.merge(full, ranking, on='Team', how='outer')
        full = pd.merge(full, defense, on='Team', how='outer')
        full = pd.merge(full, action_side, on='Team', how='outer')
        full = pd.merge(full, action_zone, on='Team', how='outer')
        full['League'] = league
        full['Season'] = season.split('_')[0]
        for attr in ['Left Side','Middle of the pitch','Right Side','Own Third','Middle Third','Opposition Third']:
            full[attr] = full[attr].apply(lambda x: round(float(x[:-1])/100,2))
        aio = pd.merge(aio, full, on=full.columns.tolist(), how='outer')

```
